```python
import requests
import json

# 配置认证和URL
API_KEY = "sk-OkeZwZeZsdY7njhDknJS1JID1GIQZ9O1Mv0bvt83XrEeo5Qn"
BASE_URL = "https://api.chatfire.cn/v1"
MODEL = "gpt-4o"

# 构造请求头和请求体
headers = {"Content-Type": "application/json", "Authorization": f"Bearer {API_KEY}"}

data = {"model": MODEL,
    "messages": [{"role": "system", "content": "你是一名系统设计师，严格按要求输出完整、结构化的系统模型方案。"},
        {"role": "user", "content": (
            "设计一个prompt，我现在需要你帮我设计一个博客系统，系统的设计应当尽量简洁，按照需求，依次输出这套系统的：用例图、系统顺序图、概念类图以及OCL合约。")}],
    "max_tokens": 3000, "temperature": 0.6}

try:
    # 发送请求到 OpenAI API
    response = requests.post(f"{BASE_URL}/chat/completions",
        headers=headers,
        data=json.dumps(data))

    # 检查响应状态码
    response.raise_for_status()

    # 解析 JSON 响应
    result = response.json()

    # 提取并打印结果
    print("=== 博客系统设计方案 ===")
    print(result["choices"][0]["message"]["content"])

except requests.exceptions.RequestException as e:
    print(f"请求失败：{e}")
except (KeyError, json.JSONDecodeError) as e:
    print(f"解析响应失败：{e}")
except Exception as e:
    print(f"发生未知错误：{e}")
```
