```python
from openai import OpenAI  # 导入SDK客户端


# 配置认证和URL
API_KEY = "sk-OkeZwZeZsdY7njhDknJS1JID1GIQZ9O1Mv0bvt83XrEeo5Qn"
BASE_URL = "https://api.chatfire.cn/v1"

# 创建客户端
client = OpenAI(
    api_key=API_KEY,
    base_url=BASE_URL
)


# 2. 构造对话消息
messages = [
    {
        "role": "system",
        "content": "你是一名系统设计师，严格按要求输出完整、结构化的系统模型方案。"
    },
    {
        "role": "user",
        "content": (
            "设计一个prompt，我现在需要你帮我设计一个博客系统，系统的设计应当尽量简洁，按照需求，依次输出这套系统的：用例图、系统顺序图、概念类图以及OCL合约。"
        )
    }
]

# 3. 调用 Chat Completions API
try:
    response = client.chat.completions.create(
        model="gpt-4o",  # 模型选择
        messages=messages,
        max_tokens=3000,  # 限制输出长度，确保复杂方案完整
        temperature=0.6   # 控制创意度（0→严谨，2→开放）
    )
    # 4. 提取并打印结果
    print("=== 博客系统设计方案 ===")
    print(response.choices[0].message.content)
except Exception as e:
    print(f"调用失败：{e}")  # 捕获网络、权限等异常
```
